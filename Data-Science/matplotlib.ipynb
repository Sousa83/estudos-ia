```python
import matplotlib.pyplot as plt
```

```python
#Criando um gráfico gerando o eixo Y e o X aleatório
#Os valores aqui são lineares, estão em uma sequência, por isso gerou uma reta
plt.plot([1, 2, 3, 4])
```

```python
#Criando um gráfico gerando o eixo Y e o X 
#O segundo array são os pontos. Como nele o valor máximo é 4, não há necessidade de Y ir além
plt.plot([1, 2, 3, 4, 5, 6], [1, 2, 3, 4, 2, 1])
```

```python
#Colocando legendas e titulo
plt.title('matplotlib')
plt.ylabel('Y')
plt.xlabel('X')
plt.plot([1, 2, 3, 4, 5, 6], [1, 2, 3, 4, 2, 1])
```
