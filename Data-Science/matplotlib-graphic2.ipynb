```python
#Gráfico de dispersão
import matplotlib.pyplot as plt
```

```python
xp1 = [1, 1, 1.5, 2, 2]
yp1 = [3, 3.5, 4, 4.5, 5]

xp2 = [3, 3, 3.5, 4, 2]
yp2 = [2, 2.5, 3, 3.5, 2]

x = [1, 2, 3, 4]
y = [2, 3, 4, 5]

plt.plot(x,y)

# "scatter()" é usado para o gráfico de dispersão.
#Este é muito usado em I.A, especificamente quando se quer resolver problemas categóricos com a mesma. Pensando nisso, o as bolinhas
#são como os dados classificados, azuiz são uma classificação X e os laranjas são uma outra classificação Y 
plt.scatter(xp1, yp1)
plt.scatter(xp2, yp2)
```
