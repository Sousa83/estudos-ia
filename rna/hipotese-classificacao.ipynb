```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#A função sigmoid é apropriada para problemas de classificação lógica, um ou outro
#Como ela nunca retorna 1 e nem 0 e sim algo entre esse valor, ela vai dizer o quanto em % 
#se parece com o A ou com B
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
nums = np.arange(10, -10, step=-1)
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(nums, sigmoid(nums))
```
