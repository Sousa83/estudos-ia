```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))
```

```python
#já temos a função de ativação (sigmoid). O cálculo da derivada parcial é y*(1-y). y nada mais é do que o resultado da sigmoid
def derivadaParcial(sigmoid):
    return sigmoid * (1-sigmoid)
```

```python
x = np.array([1, 2, 3, 4, 5])
valoresSigmoid = sigmoid(x)
dp = derivadaParcial(valoresSigmoid)
dp
```

```python
plt.plot(x, dp)
```
