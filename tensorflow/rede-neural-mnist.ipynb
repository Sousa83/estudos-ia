```python
import tensorflow as tf 
import numpy as np
import datetime
from tensorflow.keras.datasets import fashion_mnist
```

```python
#x tem as imagens, y tem as classes referentes a classificação das imagens
(xTrain, yTrain), (xTest, yTest) = fashion_mnist.load_data()
```

```python
#Verificando o conteúdo da base de treino
xTrain
```

```python
#Primeira imagem: 28px X 28px, cada elemento corresponde a cor em escala RGB (0 a 255)
xTrain[0]
```

```python
#normalizando as imagens para tornar a escala de pixel para 0 até 1, onde 0. Isso vai representar a escala de 0 a 255 das cores
xTrain = xTrain / 255
xTest = xTest / 255
```

```python
#Visualizando a primeira imagem
xTrain[0]
```

```python
#Refatorando a base para ser um vetor
xTrain.shape
```

```python
#Redimensionar todas as imagens (-1), multiplicando a altura pela largura
xTrain = xTrain.reshape(-1, 28*28)
xTest = xTest.reshape(-1, 28*28)
xTrain.shape
```

```python
#Teremos uma rede com 784 entradas, o código abaixo cria um modelo
model = tf.keras.models.Sequential()
model
```

```python
#Adicionando a primeira camada densa (oculta) com 128 nó's, com a função relu e a camada de entrada com 784 nó's
model.add(tf.keras.layers.Dense(units=128, activation='relu', input_shape=(784, )))
```

```python
#Fazendo dropout na camada densa: isso é para previnir de que aconteça sobre ajuste sobre do nó, fazemos isso zerando-o
#no caso, estou zerando (drop out) 20% dos nó's (26) serão zerados
model.add(tf.keras.layers.Dropout(0.2))
```

```python
model.add(tf.keras.layers.Dense(units=10, activation='softmax'))
```

```python
#Compilando e treinando a rede
#Adicionando um optimizador (adam) e passando como loss function e métricas a sparse_categorical..., muito utilizada em problemas categóricos diversos
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_crossentropy'])
```

```python
model.summary()
```

```python
#treinando o modelo 
model.fit(xTrain, yTrain, epochs=10)
```

```python
testLoss, testAccuracy = model.evaluate(xTest, yTest)
```
